##Equal(==)

```python
5 == 5.0
```

```python
"500mg" == "500mg"
```

```python
analgesics = ["paracetamol","ibuprofen","diclofenac"]
analgesics == analgesics
```

##Not Equal

```python
5 != 5.0
```

```python
"500mg" != "500mg"
```

```python
analgesics = ["paracetamol","ibuprofen","diclofenac"]
analgesics != analgesics
```
